### Change the names of the Columns

```python
vehicle_accident_df.columns = ['date', 'time', 'borough','zip_code','lat', 'long', 'location', 'on_street_name',
                           'cross_street_name','off_street_name','num_persons_injured','num_persons_killed',
                           'num_pedestrians_injured','num_pedestrians_killed','num_cyclist_injured','num_cyclist_killed',
                           'num_motorist_injured','num_motorist_killed','contrib_fact_one','contrib_fact_two',
                           'contrib_fact_three','contrib_fact_four','contrib_fact_five','unique_key',
                           'veh_type_1', 'veh_type_2','veh_type_3','veh_type_4', 'veh_type_5']
```

### View Changes

```python
vehicle_accident_df.head()
```

```python
vehicle_accident_df.index.name = 'id'
vehicle_accident_df.head()
```

### Get only the Columns that we want

```python
veh_accident_df = vehicle_accident_df[['date', 'time', 'borough','zip_code', 'lat','long','location','num_persons_injured',
                                       'num_persons_killed','num_pedestrians_injured','num_pedestrians_killed',
                                       'num_cyclist_injured','num_cyclist_killed','num_motorist_injured',
                                       'num_motorist_killed','contrib_fact_one','veh_type_1']]

```

### Print again to check

```python
veh_accident_df.head()
```

### Check info

```python
veh_accident_df.info()
```

### Remove rows with blanks

```python
veh_accident_df = veh_accident_df.dropna()
```

```python
veh_accident_df.info()
```

### Remove all the "straggler" data points
#### Find the top 30 in value_counts and filter out the rest

```python
veh_accident_df['veh_type_1'].value_counts()
```

### Filter out data that didn't make the top 30

```python
veh_accident_df = veh_accident_df.groupby('veh_type_1').filter(lambda x: len(x) > 588)
```

```python
veh_accident_df.info()
```

### Display value_counts after filter

```python
veh_accident_df['veh_type_1'].value_counts()
```

### Now lets group these vehicle types to match body types from ny_registration file
#### Set dict to replace values using .replace method

```python
replace_values = {'SPORT UTILITY / STATION WAGON':'SUBURBAN', 
                  'Station Wagon/Sport Utility Vehicle':'SUBURBAN','Taxi':'TAXI','VAN':'PASS_TRUCK',
                'PICK-UP TRUCK': 'PASS_TRUCK', 'Bus':'BUS','SMALL COM VEH(4 TIRES) ':'COMM_TRUCK',
            'LARGE COM VEH(6 OR MORE TIRES)':'COMM_TRUCK','LIVERY VEHICLE':'TAXI','Pick-up Truck':'PASS_TRUCK',
            'Box Truck':'COMM_TRUCK', 'MOTORCYCLE':'CYCLES','BICYCLE':'CYCLES', 'AMBULANCE':'EMERGENCY',
            'TK':'COMM_TRUCK', 'BU':'BUS', 'Van':'PASS_TRUCK', 'Bike':'CYCLES','Tractor Truck Diesel':'COMM_TRUCK',
            'FIRE TRUCK':'EMERGENCY', 'Motorcycle':'CYCLES', 'Dump':'COMM_TRUCK', 'Convertible':'CONV','Sedan':'SEDAN',  
            'UNKNOWN':'OTHER', 'PASSENGER VEHICLE':'PASSENGER_VEHICLE'    }

```

```python
new_accident_df = veh_accident_df.replace({'veh_type_1': replace_values})
```

### New accident dataframe to read into CSV as a table

```python
new_accident_df.head()
```

### Drop the columns not needed

```python
new_accident_df = new_accident_df[['date', 'time', 'borough','zip_code','num_persons_injured',
                                       'num_persons_killed','num_pedestrians_injured','num_pedestrians_killed',
                                       'num_cyclist_injured','num_cyclist_killed','num_motorist_injured',
                                       'num_motorist_killed','veh_type_1']]
new_accident_df.head()
```

```python
new_accident_df.info()
```

```python
new_accident_df.to_csv("nyc_collisions.csv")
```

### Grouped by zipcode and veh_type_1

```python
accident_by_zip = new_accident_df.groupby(['zip_code','veh_type_1'])['num_persons_injured','num_persons_killed','num_pedestrians_injured','num_pedestrians_killed',
                                            'num_cyclist_injured','num_cyclist_killed','num_motorist_injured',
                                            'num_motorist_killed'].sum()
```
